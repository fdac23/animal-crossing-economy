```python
def sorted_bar(_df, _x: str, _y: str):
    s_df = _df.groupby([_x])[_y].aggregate(np.median).reset_index().sort_values(_y)
    sns.barplot(data=s_df, x=_x, y=_y)
    plt.show()
    
def sorted_bar_pal(_df, _x: str, _y: str, pal):
    s_df = _df[[_x, _y]].sort_values(_y, ascending=False)
    sns.barplot(data=s_df, x=_x, y=_y, palette=pal)
    plt.show()
```

```python
sns.set_style("darkgrid")
sorted_bar(trees, "Name", "Sell")
sorted_bar_pal(flowers, "Color 1", "Sell", f_palet)
plt.xticks(rotation=45)
sorted_bar(fromTree, "Name", "Sell")
```
